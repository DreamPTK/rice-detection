## Load Package

```python
import datacube
from dream_river.plotimg import img_OnMap
```

## Connect to datacube

```python
dc = datacube.Datacube(app='Imagery_on_web_map')
```

## Set parameter

```python
# Define product and red/green/blue bands in the given product
product = 'ls8_collections_sr_scene'
RGB = ('red', 'green', 'blue')

# Region and time of interest
query = dict(
    lat=(15.682, 15.728),
    lon=(100.179, 100.247),
    time='2016-01-23',
)
```

## Find dataset form datacube

```python
dss = dc.find_datasets(product=product, **query)
print(f"Found {len(dss)} datasets")
```

## Load data

```python
ds = dc.load(product=product,             
             datasets=dss,                
             measurements=RGB,           
             group_by='solar_day',        
             output_crs='EPSG:32647',      
             resolution=(-30, 30),     
             resampling='bilinear', )  
ds
```

## Display satellite imagery on interactive map

```python
img_OnMap(dss, ds)
```
